## Fibonacci revisited

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def fibonacci(n):
    """
    Returns the Fibonacci sequence from 0 to n
    :param n:
    :return:
    """
    if n in {0,1}:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)
```

```python
# x and y coordinates
x = np.arange(1,20)
y = np.array([fibonacci(i) for i in x])

fig, ax = plt.subplots()
ax.scatter(x,y)
ax.set(xlabel="n", ylabel="fib(n)", title="The Fibonacci sequence")
```
